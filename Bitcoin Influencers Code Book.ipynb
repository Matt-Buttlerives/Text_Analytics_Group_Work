```python
#### Text Pre-processing ####
```

```python
##### 1: Remove Punctuations
# Defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

# Storing the puntuation free text
mmdf['punctuation_free']= mmdf['Text'].apply(lambda x:remove_punctuation(x))
rvdf['punctuation_free']= rvdf['Text'].apply(lambda x:remove_punctuation(x))
aadf['punctuation_free']= aadf['Text'].apply(lambda x:remove_punctuation(x))
tddf['punctuation_free']= tddf['Text'].apply(lambda x:remove_punctuation(x))
apdf['punctuation_free']= apdf['Text'].apply(lambda x:remove_punctuation(x))

```

```python
##### 2: Lowering Text
mmdf['msg_lower']= mmdf['punctuation_free'].apply(lambda x: x.lower())
rvdf['msg_lower']= rvdf['punctuation_free'].apply(lambda x: x.lower())
aadf['msg_lower']= aadf['punctuation_free'].apply(lambda x: x.lower())
tddf['msg_lower']= tddf['punctuation_free'].apply(lambda x: x.lower())
apdf['msg_lower']= apdf['punctuation_free'].apply(lambda x: x.lower())


##### 3: Tokenization
mmdf['msg_tokenized']= mmdf['msg_lower'].apply(lambda x: word_tokenize(x))
rvdf['msg_tokenized']= rvdf['msg_lower'].apply(lambda x: word_tokenize(x))
aadf['msg_tokenized']= aadf['msg_lower'].apply(lambda x: word_tokenize(x))
tddf['msg_tokenized']= tddf['msg_lower'].apply(lambda x: word_tokenize(x))
apdf['msg_tokenized']= apdf['msg_lower'].apply(lambda x: word_tokenize(x))
```

```python
# Stop words present in the library
import nltk
stopwords = nltk.corpus.stopwords.words('english')

# Defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output

# Applying the function
mmdf['no_stopwords']= mmdf['msg_tokenized'].apply(lambda x:remove_stopwords(x))
rvdf['no_stopwords']= rvdf['msg_tokenized'].apply(lambda x:remove_stopwords(x))
aadf['no_stopwords']= aadf['msg_tokenized'].apply(lambda x:remove_stopwords(x))
tddf['no_stopwords']= tddf['msg_tokenized'].apply(lambda x:remove_stopwords(x))
apdf['no_stopwords']= apdf['msg_tokenized'].apply(lambda x:remove_stopwords(x))

##### 5: Lemmatization 
# Defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()


# Defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text

mmdf['msg_lemmatized']=mmdf['no_stopwords'].apply(lambda x:lemmatizer(x))
rvdf['msg_lemmatized']=rvdf['no_stopwords'].apply(lambda x:lemmatizer(x))
aadf['msg_lemmatized']=aadf['no_stopwords'].apply(lambda x:lemmatizer(x))
tddf['msg_lemmatized']=tddf['no_stopwords'].apply(lambda x:lemmatizer(x))
apdf['msg_lemmatized']=apdf['no_stopwords'].apply(lambda x:lemmatizer(x))

clean_posts_mmdf = mmdf.msg_lemmatized
keywords_mmdf = mmdf.msg_lemmatized.copy()
clean_posts_rvdf = rvdf.msg_lemmatized
keywords_rvdf = rvdf.msg_lemmatized.copy()
clean_posts_aadf = aadf.msg_lemmatized
keywords_aadf = aadf.msg_lemmatized.copy()
clean_posts_tddf = tddf.msg_lemmatized
keywords_tddf = tddf.msg_lemmatized.copy()
clean_posts_apdf = apdf.msg_lemmatized
keywords_apdf = apdf.msg_lemmatized.copy()
```

```python
eldc = pd.read_csv('@Elon_2021_bitcoin_tweets_users.csv', names = ['Datetime','Tweet Id','Text', 'Username', 'url'])
##### 1: Remove Punctuations
# Defining the function to remove punctuation
def remove_punctuation(text):
    punctuationfree="".join([i for i in text if i not in string.punctuation])
    return punctuationfree

# Storing the puntuation free text
eldc['punctuation_free']= eldc['Text'].apply(lambda x:remove_punctuation(x))
```

```python
##### 2: Lowering Text
eldc['msg_lower']= eldc['punctuation_free'].apply(lambda x: x.lower())


##### 3: Tokenization
eldc['msg_tokenized']= eldc['msg_lower'].apply(lambda x: word_tokenize(x))
```

```python
# Stop words present in the library
import nltk
stopwords = nltk.corpus.stopwords.words('english')

# Defining the function to remove stopwords from tokenized text
def remove_stopwords(text):
    output= [i for i in text if i not in stopwords]
    return output

# Applying the function
eldc['no_stopwords']= eldc['msg_tokenized'].apply(lambda x:remove_stopwords(x))


##### 5: Lemmatization 
# Defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()


# Defining the function for lemmatization
def lemmatizer(text):
    lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
    return lemm_text

eldc['msg_lemmatized']=eldc['no_stopwords'].apply(lambda x:lemmatizer(x))


clean_posts_eldc = eldc.msg_lemmatized
keywords_eldc =eldc.msg_lemmatized.copy()
```

```python
#defining an object for frequency distribution of words   

fdist = FreqDist(words)
coin = pd.DataFrame.from_dict(fdist, orient='index')
coin.columns = ['Frequency']
coin.index.name = 'Token'
coin = coin.reset_index()
#sorting by frequency, descending
coin = coin.sort_values(by=['Frequency'], ascending=False)
#obtaining the top 5 coins
top_5 = coin.head(5)
#generating list of top 5 coins

top_5_coins = top_5['Token'].tolist()
top_5_coins = list(dict.fromkeys(top_5_coins))
#deleting tokens not in the top 5 coins
for i in range(len(clean_posts_eldc)):
    clean_posts_eldc[i] = [j for j in clean_posts_eldc[i] if j in top_5_coins]
#deleting rows with empty list
top_coins_posts_eldc = list(filter(None, clean_posts_eldc))  
#getting all combinations of pairs
all_pairs = list(itertools.product(top_5_coins, repeat=2))
#counting the number of cooccurences of each pair
cooccurence_eldc = {pair: len([x for x in top_coins_posts_eldc if set(pair) <= set(x)]) for pair in all_pairs}
#creating an empty dataframe to contain tokens used in conjunction with each coin
keywords_count = pd.DataFrame(columns=['Token','Word_List'])
for i in range(len(top_5_coins)):
    keywords_count = keywords_count.append({'Token':top_5_coins[i],
                                            'Word_List':[top_5_coins[i],]},
                                           ignore_index=True)
```

```python
# creating an empty list to contain tokens used for all cars
all_tokens = []

#assigning tokens to coins it mentioned together with and creating a list of tokens in every posts
for i in range(len(keywords_eldc)):
    for j in range(len(top_5_coins)):
        if top_5_coins[j] in keywords_eldc[i]:
            keywords_count.iloc[j,1] += list(dict.fromkeys(keywords_eldc[i]))
    all_tokens.extend(list(dict.fromkeys(keywords_eldc[i])))
```
